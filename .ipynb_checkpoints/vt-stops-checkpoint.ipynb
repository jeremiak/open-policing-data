```python
%matplotlib inline
from datetime import date
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vt = pd.read_csv('./data/VT-clean.csv', low_memory=False)
```

```python
def date_from_string(s):
    spl = s.split('-')
    return date(int(spl[0]), int(spl[1]), int(spl[2]))

def days_from_new_year(d):
    ny = date(d.year, 1, 1)
    diff = d - ny
    return diff.days

def minutes_from_new_year(d, m):
    days = days_from_new_year(d)
    min_per_day = 60 * 24
    return (days * min_per_day) + m

def minutes_from_string(s):
    t = s.split(':')
    return (int(t[0]) * 60) + int(t[1])
```

```python
def process_row(row):
    stop_date = date_from_string(row['stop_date'])
    stop_minutes = minutes_from_string(row['stop_time']) # minutes into the day, midnight is 0
    mins = minutes_from_new_year(stop_date, stop_minutes)
    return mins
```

```python
vt['min_from_new_year'] = vt.apply(lambda row: process_row(row),axis=1)
```

```python
mins = vt[['id', 'stop_date', 'stop_time', 'stop_outcome', 'min_from_new_year']]
```

```python
outcomes = mins['stop_outcome'].unique()
```

```python
for o in outcomes:
    dd = mins[mins['stop_outcome'] == o]
    if len(dd) != 0:
        c = dd['min_from_new_year'].plot.hist(title=o)
        c.set_xlabel("Minutes since the New Year")
        plt.show()

# by seasonality
# seems like this shows some pretty clear seasonality to the types of things that happen
# after a stop. is the verbal warning pattern indicitative of quotas or holiday spirit?
```

```python
for o in outcomes:
    dd = mins[mins['stop_outcome'] == o]
    if len(dd) != 0:
        c = dd['stop_time'].apply(lambda s: minutes_from_string(s)).plot.hist(title=o)
        c.set_xlabel('Minutes since midnight (~1430 is 1159PM)')
        plt.show()

# by time of day
# when are shift changes?
# would be good to see this as regular, human readable hours instead of of minutes since midnight
```
